Standardize ride_status (e.g., completed, cancelled) and payment_method.

```python
# Define mapping dictionaries
status_map = {
    "Completed": ["completed", "complete", "done", "finished", "success"],
    "Cancelled": ["canceled", "cancelled", "cancel", "cancelled_by_driver", "cancelled_by_user"],
}

payment_map = {
    "UPI": ["upi", "upi_payment", "gpay", "phonepe", "paytm"],  
}

# Standardization helper function
def standardize(value, mapping):
    value = str(value).strip().lower()
    for standard, variants in mapping.items():
        if value in variants:
            return standard
    return "other"  # fallback for unexpected values

# Apply mappings
df["ride_status"] = df["ride_status"].apply(lambda x: standardize(x, status_map))
df["payment_method"] = df["payment_method"].apply(lambda x: standardize(x, payment_map))

# Preview the cleaned data
print(df.head(5))
             
```

Detect outliers in ride_charge or total_fare.


```python
# Choose the column (can use 'ride_charge' or 'total_fare')
col = 'total_fare'

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

print(f"Number of outliers in {col}: {len(outliers)}")
print(outliers[[col]].head())
```

Predict fare (total_fare) based on distance, duration, and service type.


```python
# importing scikit_learn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = df.dropna(subset=['total_fare'])

x = df[['distance', 'duration']]
y = pd.get_dummies(pd.concat([x, df['services']], axis=1), drop_first=True)
y = df['total_fare']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
print('Example Prediction:', y_pred[:5])
```

Use correlation analysis (numpy.corrcoef) to see the relation between distance, duration, and total_fare.


```python
data = df[['distance', 'duration', 'total_fare']]
corr_matrix = np.corrcoef(data.T)

corr_df = pd.DataFrame(corr_matrix, 
                       index=['distance', 'duration', 'total_fare'], 
                       columns=['distance', 'duration', 'total_fare'])
print(corr_df)


# 1  perfect positive correlation
# 0  no correlation
# -1  perfect negative correlation
```
